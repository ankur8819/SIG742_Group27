```python


import ast
import numpy as np
import pandas as pd
from scipy import sparse
from sklearn.preprocessing import MultiLabelBinarizer, normalize
from sklearn.neighbors import NearestNeighbors

# ---------- 1) Load & preprocess data ----------
def load_and_preprocess(business_csv="meta-review-business.csv", review_csv="review.csv"):
    # Load
    business_df = pd.read_csv(business_csv)
    review_df = pd.read_csv(review_csv)
    
    # Convert review "time" (likely ms since epoch) to datetime
    try:
        review_df['time'] = pd.to_datetime(review_df['time'], unit='ms')
    except Exception:
        # fallback (if already seconds)
        review_df['time'] = pd.to_datetime(review_df['time'], unit='s', errors='coerce')
    
    # Normalize gmap_id to string
    business_df['gmap_id'] = business_df['gmap_id'].astype(str)
    review_df['gmap_id'] = review_df['gmap_id'].astype(str)
    
    # Parse categories (category may be stored as a string-repr list)
    def safe_parse_list(x):
        if pd.isna(x):
            return []
        if isinstance(x, list):
            return x
        try:
            # Some rows use bracket-list strings, others may be simple text
            parsed = ast.literal_eval(x)
            # if parsed is a string, wrap it
            if isinstance(parsed, str):
                return [parsed]
            return parsed
        except Exception:
            # fallback: split on comma if looks like "A, B"
            if isinstance(x, str):
                if ',' in x:
                    return [s.strip().strip("'\"") for s in x.split(',') if s.strip()]
                else:
                    return [x.strip()]
            return []
    
    business_df['categories_parsed'] = business_df['category'].apply(safe_parse_list)
    
    # Aggregate reviews: if there are duplicates (same user_id, same gmap_id),
    # compute the mean rating or take the latest rating. We'll use mean here.
    # Also drop reviews with missing rating
    reviews = review_df.dropna(subset=['rating']).copy()
    # Ensure user_id is string (some are floats/scientific)
    reviews['user_id'] = reviews['user_id'].astype(str)
    # Aggregate
    reviews_agg = reviews.groupby(['user_id', 'gmap_id'], as_index=False).agg({
        'rating': 'mean',  # if multiple, average
        'time': 'max'      # keep the latest time (optional)
    })
    
    return business_df, reviews_agg

# ---------- 2) Build item-user matrix ----------
def build_item_user_matrix(reviews, business_df,
                           min_user_ratings=5, min_item_ratings=10,
                           keep_businesses_without_reviews=False):
    """
    reviews: DataFrame with columns ['user_id','gmap_id','rating']
    business_df: DataFrame with 'gmap_id'
    returns:
      - item_user_matrix (scipy.sparse.csr_matrix) shape = (n_items, n_users)
      - item_index_to_gmap (list), gmap_to_item_index (dict)
      - user_index_to_id (list), user_id_to_index (dict)
      - filtered_business_df (DataFrame) aligned to item indices
    """
    # Count user and item activity
    user_counts = reviews['user_id'].value_counts()
    item_counts = reviews['gmap_id'].value_counts()
    
    # Keep only users and items with enough activity
    keep_users = set(user_counts[user_counts >= min_user_ratings].index)
    keep_items = set(item_counts[item_counts >= min_item_ratings].index)
    
    reviews_f = reviews[reviews['user_id'].isin(keep_users) & reviews['gmap_id'].isin(keep_items)].copy()
    
    # Build index mapping for users and items (only those kept)
    user_ids = sorted(reviews_f['user_id'].unique())
    item_ids = sorted(reviews_f['gmap_id'].unique())
    user_id_to_index = {uid:i for i,uid in enumerate(user_ids)}
    item_id_to_index = {iid:i for i,iid in enumerate(item_ids)}
    
    # Build COO arrays
    rows = reviews_f['gmap_id'].map(item_id_to_index).to_numpy()
    cols = reviews_f['user_id'].map(user_id_to_index).to_numpy()
    data = reviews_f['rating'].to_numpy(dtype=float)
    
    # Create sparse matrix (items x users)
    item_user_coo = sparse.coo_matrix((data, (rows, cols)), shape=(len(item_ids), len(user_ids)))
    item_user = item_user_coo.tocsr()
    
    # Optionally include businesses that were filtered out (no/low reviews) into business_df
    filtered_business_df = business_df[business_df['gmap_id'].isin(item_ids)].copy()
    filtered_business_df = filtered_business_df.set_index('gmap_id').loc[item_ids].reset_index()
    
    return item_user, item_ids, user_ids, item_id_to_index, user_id_to_index, filtered_business_df

# ---------- 3) Build KNN graph (top-K neighbors per item) ----------
def build_item_knn_graph(item_user_matrix, n_neighbors=50, metric='cosine', n_jobs=1):
    """
    Build top-K nearest neighbors for each item using item vectors (rows of item_user_matrix).
    Returns a sparse similarity matrix S (CSR) where S[i,j] = similarity between item i and j (nonzero only for top-K neighbors).
    """
    # Use NearestNeighbors with metric='cosine' (returns distances in [0,2], where 0 is identical)
    nn = NearestNeighbors(n_neighbors=n_neighbors+1, metric=metric, algorithm='brute', n_jobs=n_jobs)
    nn.fit(item_user_matrix)  # items are rows
    
    distances, neighbors = nn.kneighbors(item_user_matrix, return_distance=True)
    # distances shape = (n_items, n_neighbors+1) - first neighbor is the item itself (distance 0)
    n_items = item_user_matrix.shape[0]
    
    rows = []
    cols = []
    data = []
    
    for i in range(n_items):
        for dist, nbr in zip(distances[i,1:], neighbors[i,1:]):  # skip self (index 0)
            sim = 1.0 - dist  # cosine similarity approximate (since dist = 1 - cosine) but sklearn returns distance=cosine
            if sim > 0:
                rows.append(i)
                cols.append(nbr)
                data.append(sim)
    # Build sparse matrix item x item
    S = sparse.csr_matrix((data, (rows, cols)), shape=(n_items, n_items))
    # Optionally normalize rows (so predictions are weighted averages)
    S = normalize(S, norm='l1', axis=1)  # row-stochastic: row sums = 1
    return S

# ---------- 4) Content (category) similarity ----------
def build_category_similarity(business_df_filtered, top_k=50):
    """
    business_df_filtered must be aligned to item_ids (order matters).
    Returns a row-normalized sparse similarity matrix C (items x items) built from categories.
    """
    mlb = MultiLabelBinarizer(sparse_output=True)
    cat_matrix = mlb.fit_transform(business_df_filtered['categories_parsed'].apply(lambda x: x if isinstance(x, (list,tuple)) else []))
    # cat_matrix is sparse (items x n_categories)
    # Compute cosine similarity via NearestNeighbors to get top_k neighbors for each item
    nn = NearestNeighbors(n_neighbors=min(top_k+1, cat_matrix.shape[0]), metric='cosine', algorithm='brute')
    nn.fit(cat_matrix)
    distances, neighbors = nn.kneighbors(cat_matrix, return_distance=True)
    
    rows, cols, data = [], [], []
    for i in range(cat_matrix.shape[0]):
        for dist, nbr in zip(distances[i,1:], neighbors[i,1:]):
            sim = 1.0 - dist
            if sim > 0:
                rows.append(i); cols.append(nbr); data.append(sim)
    C = sparse.csr_matrix((data, (rows, cols)), shape=(cat_matrix.shape[0], cat_matrix.shape[0]))
    C = normalize(C, norm='l1', axis=1)
    return C

# ---------- 5) Combine similarity graphs (hybrid) ----------
def combine_similarity_graphs(S_collab, S_content, alpha=0.7):
    """
    alpha: weight for collaborative similarity. (1-alpha) weight for content similarity.
    S_* are row-normalized sparse matrices (CSR).
    Returns row-normalized combined S.
    """
    # Convert to same sparse format and combine
    S = alpha * S_collab + (1.0 - alpha) * S_content
    S = normalize(S, norm='l1', axis=1)
    return S

# ---------- 6) Recommendation functions ----------
def recommend_similar_businesses(gmap_id, item_ids, business_df_filtered, S_combined,
                                 top_n=10, exclude_self=True):
    """
    Recommend businesses similar to the given gmap_id using the combined similarity graph.
    """
    if gmap_id not in item_ids:
        return pd.DataFrame()  # fallback: empty (business not in collab space)
    idx = item_ids.index(gmap_id)
    row = S_combined.getrow(idx).toarray().ravel()
    # rank
    top_idx = np.argsort(row)[::-1]
    if exclude_self:
        top_idx = top_idx[top_idx != idx]
    top_idx = top_idx[:top_n]
    recs = business_df_filtered.iloc[top_idx].copy()
    recs['similarity_score'] = row[top_idx]
    return recs[['gmap_id','name','address','categories_parsed','avg_rating','num_of_reviews','similarity_score']]

def predict_for_user(user_id, user_id_to_index, item_ids, S_combined, item_user_matrix, top_n=10):
    """
    Predict scores for a user using item-based CF:
      pred_item_scores = S_combined dot user_ratings_vector
    Returns top_n recommendations (gmap_id, predicted_score).
    """
    if user_id not in user_id_to_index:
        return pd.DataFrame()  # cold-start user fallback outside this filtered set
    uidx = user_id_to_index[user_id]
    # user ratings vector aligned to columns of item_user_matrix (items x users)
    user_ratings = item_user_matrix[:, uidx].toarray().ravel()  # shape n_items
    # compute predicted scores
    pred_scores = S_combined.dot(user_ratings)
    # Do not recommend items the user already rated
    already_rated = user_ratings > 0
    pred_scores[already_rated] = -np.inf
    top_idx = np.argsort(pred_scores)[::-1][:top_n]
    res = pd.DataFrame({
        'gmap_id': [item_ids[i] for i in top_idx],
        'predicted_score': pred_scores[top_idx]
    })
    return res

# ---------- 7) Putting it all together ----------
def build_and_run_pipeline(business_csv="meta-review-business.csv", review_csv="review.csv",
                           min_user_ratings=5, min_item_ratings=10,
                           knn_k=50, cat_k=50, hybrid_alpha=0.7, n_jobs=1):
    business_df, reviews_agg = load_and_preprocess(business_csv, review_csv)
    print("Loaded: businesses =", len(business_df), "reviews (agg rows) =", len(reviews_agg))
    item_user, item_ids, user_ids, item_id_to_index, user_id_to_index, business_filtered = build_item_user_matrix(
        reviews_agg, business_df, min_user_ratings=min_user_ratings, min_item_ratings=min_item_ratings
    )
    print("Filtered item_user matrix:", item_user.shape, "items kept:", len(item_ids), "users kept:", len(user_ids))
    
    # Build collaborative KNN graph
    print("Building collaborative KNN graph (k=%d)..." % knn_k)
    S_collab = build_item_knn_graph(item_user, n_neighbors=knn_k, n_jobs=n_jobs)
    
    # Build category similarity
    print("Building category similarity (k=%d)..." % cat_k)
    S_cat = build_category_similarity(business_filtered, top_k=cat_k)
    
    # Combine
    # print("Combining graphs alpha=", hybrid_alpha)
    # S_combined = combine_similarity_graphs(S_collab, S_cat, alpha=hybrid_alpha)
    
    # return {
    #     'business_df': business_df,
    #     'reviews_agg': reviews_agg,
    #     'item_user_matrix': item_user,
    #     'item_ids': item_ids,
    #     'user_ids': user_ids,
    #     'item_id_to_index': item_id_to_index,
    #     'user_id_to_index': user_id_to_index,
    #     'business_filtered': business_filtered,
    #     'S_collab': S_collab,
    #     'S_cat': S_cat,
    #     'S_combined': S_combined
    # }


```
